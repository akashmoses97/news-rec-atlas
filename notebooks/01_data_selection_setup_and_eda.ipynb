(C) Dataset Selection
======================

Selected Dataset: MIND (Microsoft News Recommendation Dataset)
---------------------------------------------------------------

After comparative evaluation, I have chosen the **MIND dataset** for this semester-long project.

The selection is based on methodological reasoning, feasibility, industry relevance, alignment with course techniques, and the ability to extend into advanced research directions.

<br>

**Reasons:**
--------------

1. **Alignment with Course Techniques**

MIND directly supports multiple core course topics:

- *Graph Mining*: The dataset forms a user-news bipartite interaction graph suitable for centrality analysis, graph construction, and community detection.
- *Frequent Itemsets*: Co-clicked articles within impressions enable association rule mining.
- *Clustering*: User segmentation based on interaction patterns and content preferences.
- *Text Mining*: News titles and abstracts allow feature extraction, TF-IDF modeling, and embedding-based representations.
- *Anomaly Detection*: Abnormal click behavior and engagement bursts can be modeled.

Unlike a purely graph-based dataset, MIND allows simultaneous exploration of **graph, transactional, and textual mining techniques**, which covers more of course techniques.


2. **Support for Beyond-Course Techniques**

MIND enables exploration of advanced techniques not covered in lecture:

- *Transformer-based sequential recommendation models*
- *LLM-based text embeddings for content-aware recommendation*
- *Fairness-aware ranking and exposure bias correction*
- *Temporal modeling of user engagement patterns*

This flexibility ensures the project can extend into modern recommender system research while being grounded in core data mining methods.


3. **Industry Relevance**

MIND reflects real-world recommender system challenges faced by:

- News platforms (Microsoft News, Google News)
- Social media feeds
- Streaming and content recommendation systems
- E-commerce personalization engines

The dataset captures key industrial problems such as:

- *Implicit feedback modeling (click data)*
- *Cold-start user and item challenges*
- *Popularity and exposure bias*
- *Large-scale ranking optimization*

Studying MIND would provide practical insight into industry-scale recommendation systems, making the project relevant to applied machine learning and data science roles.

4. **Data Structure and Analytical Depth**


MIND provides:

- Large-scale interaction logs (~1M impressions)
- Rich metadata (categories, titles, abstracts)
- Predefined train/validation splits

The combination of structured click logs and textual content allows meaningful EDA, as per the various points stated in point 3.

This structure naturally supports hypothesis generation and research question formulation.

5. **Algorithmic Feasibility**

The dataset is:

- Computationally manageable on a single machine
- Suitable for Apriori with appropriate thresholds
- Feasible for sequential modeling without distributed infrastructure
- Well-supported by existing open-source implementations


6. **Bias and Ethical Research Potential**


MIND allows structured investigation of:

- *Popularity bias*
- *Position/exposure bias*
- *Feedback loops in recommendation systems*
- *Cold-start disadvantages*

Ethically, it presents low direct harm risk, as it contains no sensitive demographic attributes.

---------------------------------------------------------------------

**Trade-offs:**
----------

- Click labels reflect implicit feedback and may not represent true user preference.
- Lack of explicit demographic data limits certain fairness analyses.
- News domain may introduce temporal concept drift.

However, these limitations are manageable within the scope of the project.

**Yelp**, while rich and realistic, introduces heavier preprocessing requirements due to JSON structure and noisy text. It also raises stronger ethical concerns related to reputational interpretation of real businesses. Although strong, it requires more engineering effort compared with MIND.

**BTS (Tie Strength Prediction)** is highly specialized toward graph learning and link prediction. While excellent for advanced GNN experimentation, it lacks textual and transactional diversity.

In contrast, **MIND provides the broadest methodological coverage, strong industry relevance, lower preprocessing burden, research extensibility, and balanced ethical considerations**, making it the most suitable dataset for this semester-long project.

```python
# ============================================================
# MIND Dataset – EXPLORATORY DATA ANALYSIS (EDA)
# ============================================================
#
# This EDA covers:
#   1. Data basics (dimensions, unique users, unique items)
#   2. Data cleaning and integrity checks (missingness, duplicates)
#   3. Basket size and user activity distributions
#   4. Click imbalance and frequency of top items (long-tail behavior)
#   5. Interaction and co-occurrence sparsity
#   6. Category and exposure bias diagnostics
#   7. Temporal interaction patterns
#   8. Correlation of user-level behavioral features
#   9. Structural validation tests
#
# NOTE: Each point header has the reason for the Algorithmic decision made
# ============================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
from itertools import combinations

sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)
np.random.seed(42)

# ------------------------------------------------------------
# 1. DATA BASICS
# ------------------------------------------------------------
# Gives dataset size, number of users, and number of items
# is critical for assessing scalability and sparsity.

BEHAVIORS_PATH = "MINDsmall_train/behaviors.tsv"
NEWS_PATH = "MINDsmall_train/news.tsv"

behaviors = pd.read_csv(
    BEHAVIORS_PATH,
    sep="\t",
    header=None,
    names=["impression_id", "user_id", "time", "history", "impressions"]
)

news = pd.read_csv(
    NEWS_PATH,
    sep="\t",
    header=None,
    names=["news_id", "category", "subcategory", "title", "abstract",
           "url", "title_entities", "abstract_entities"]
)

print("Behaviors shape:", behaviors.shape)
print("News shape:", news.shape)
print("Unique users:", behaviors["user_id"].nunique())
print("Unique news articles:", news["news_id"].nunique())

# ------------------------------------------------------------
# 2. DATA CLEANING & INTEGRITY CHECKS
# ------------------------------------------------------------
# Real-world log data often contains missing values or duplicates.
# These must be inspected before modeling.

print("\nMissing values (behaviors):")
print(behaviors.isnull().sum())

print("\nMissing values (news):")
print(news.isnull().sum())

print("\nDuplicate impression IDs:",
      behaviors["impression_id"].duplicated().sum())

print("Duplicate news IDs:",
      news["news_id"].duplicated().sum())

# ------------------------------------------------------------
# 3. DISTRIBUTION OF BASKET SIZES & USER ACTIVITY DISTRIBUTIONS
# ------------------------------------------------------------
# Each impression contains multiple candidate news items.
# This is equivalent to transaction basket size in retail datasets.

behaviors["num_candidates"] = behaviors["impressions"].apply(
    lambda x: len(str(x).split())
)

sns.histplot(behaviors["num_candidates"], bins=30)
plt.title("Distribution of Impression Sizes")
plt.xlabel("Number of Candidate News per Impression")
plt.show()

print("Mean candidates per impression:",
      round(behaviors["num_candidates"].mean(), 2))

# Cold-start users are a major real-world issue in recommendation systems.

user_activity = behaviors.groupby("user_id").size()

sns.histplot(user_activity, bins=50)
plt.title("User Activity Distribution")
plt.xlabel("Impressions per User")
plt.show()

print("Median impressions per user:",
      user_activity.median())

cold_users = (user_activity == 1).sum()
print("Cold-start users:", cold_users)
print("Cold-start percentage:",
      round(cold_users / len(user_activity), 4))

# ------------------------------------------------------------
# 4. FREQUENCY OF TOP ITEMS & CLASS IMBALANCE
# ------------------------------------------------------------
# Click logs are implicit feedback and typically highly imbalanced.
# Also identifies News popularity (long-tail behavior)]

clicked_news = []
all_labels = []

for row in behaviors["impressions"]:
    for item in row.split():
        news_id, label = item.split("-")
        all_labels.append(label)
        if label == "1":
            clicked_news.append(news_id)

label_counts = Counter(all_labels)

sns.barplot(x=list(label_counts.keys()),
            y=list(label_counts.values()))
plt.title("Click vs Non-Click Distribution")
plt.show()

print("Click ratio:",
      round(label_counts["1"] / sum(label_counts.values()), 4))

news_popularity = Counter(clicked_news)

sns.histplot(list(news_popularity.values()), bins=50)
plt.title("News Popularity Distribution (Long-Tail Analysis)")
plt.xlabel("Number of Clicks")
plt.show()

print("Top 5 most clicked articles:",
      news_popularity.most_common(5))

# ------------------------------------------------------------
# 5. SPARSITY ANALYSIS
# ------------------------------------------------------------
# Sparse interaction matrices motivate embedding-based approaches.
# User–item interaction sparsity and Item co-occurrence sparsity are calculated
# Determines feasibility of frequent itemset mining.

num_users = behaviors["user_id"].nunique()
num_news = news["news_id"].nunique()
num_interactions = len(clicked_news)

interaction_sparsity = 1 - (
    num_interactions / (num_users * num_news)
)

print("\nUser–News Interaction Sparsity:",
      round(interaction_sparsity, 6))

# Co-occurrence is computed across impressions to estimate structural sparsity.
# Full enumeration is used here for exact measurement (dataset is manageable in size).

co_occurrence_counts = Counter()

for row in behaviors["impressions"]:
    items = [item.split("-")[0] for item in row.split()]
    for pair in combinations(items, 2):
        co_occurrence_counts[tuple(sorted(pair))] += 1

observed_pairs = len(co_occurrence_counts)
possible_pairs = (num_news * (num_news - 1)) / 2

co_occurrence_sparsity = 1 - (observed_pairs / possible_pairs)

print("Item co-occurrence sparsity:",
      round(co_occurrence_sparsity, 6))

# ------------------------------------------------------------
# 6. BIAS DIAGNOSTICS
# ------------------------------------------------------------
# Category imbalance may introduce structural recommendation bias.
# Click position bias can come up with items shown earlier.

sns.countplot(
    data=news,
    y="category",
    order=news["category"].value_counts().index
)
plt.title("News Category Distribution")
plt.show()

position_clicks = []

for row in behaviors["impressions"]:
    for idx, item in enumerate(row.split()):
        news_id, label = item.split("-")
        if label == "1":
            position_clicks.append(idx)

sns.histplot(position_clicks, bins=30)
plt.title("Click Position Distribution")
plt.xlabel("Position in Impression")
plt.show()

# ------------------------------------------------------------
# 7. TEMPORAL GAPS BETWEEN TRANSACTIONS
# ------------------------------------------------------------
# This supports investigation of sequential modeling techniques.

behaviors["time"] = pd.to_datetime(behaviors["time"])
behaviors = behaviors.sort_values(["user_id", "time"])

behaviors["prev_time"] = behaviors.groupby("user_id")["time"].shift(1)
behaviors["time_gap_hours"] = (
    behaviors["time"] - behaviors["prev_time"]
).dt.total_seconds() / 3600

sns.histplot(
    behaviors["time_gap_hours"].dropna(),
    bins=50
)
plt.title("Distribution of Time Gaps Between User Interactions (Hours)")
plt.xlabel("Hours")
plt.show()

# ------------------------------------------------------------
# 8. CORRELATION ANALYSIS (User-Level Structural Features)
# ------------------------------------------------------------
# This examines relationships between user engagement intensity,
# browsing history depth and temporal behavior.
# This helps identify structural dependencies that may influence
# modeling choices.

# Computing history length
behaviors["history_length"] = behaviors["history"].fillna("").apply(
    lambda x: len(x.split()) if x else 0
)

# Aggregating user-level features
# These features capture engagement depth, exposure complexity,
# and temporal behavior — core dimensions of user interaction.
user_features = behaviors.groupby("user_id").agg({
    "history_length": "mean",
    "num_candidates": "mean",
    "time_gap_hours": "mean"
}).reset_index()

user_features["activity_count"] = user_activity.values

corr_matrix = user_features.drop(columns=["user_id"]).corr()

# Plotting heatmap
plt.figure(figsize=(8,6))
sns.heatmap(
    corr_matrix,
    annot=True,
    cmap="coolwarm",
    fmt=".2f",
    linewidths=0.5
)
plt.title("Correlation Heatmap of User-Level Behavioral Features")
plt.show()

# ------------------------------------------------------------
# NOTE : Initial observations motivating external techniques
# (this content is present in markdown)
# ------------------------------------------------------------

# ============================================================
# 9. EDA VALIDATION TESTS
# ============================================================
# These tests validate structural integrity, consistency,
# and statistical assumptions of the dataset.
# ============================================================

print("\nRunning validation checks...\n")


# 1. Every impression must contain at least one candidate
assert (behaviors["num_candidates"] > 0).all(), \
    "Found impression with zero candidates."

print("TEST-1 pass : All impressions contain at least one candidate.")


# 2. Click ratio sanity check (should be low)
click_ratio = label_counts["1"] / sum(label_counts.values())

assert click_ratio < 0.5, \
    "Click ratio unexpectedly high; check parsing logic."

print("TEST-2 pass : Click ratio is realistically imbalanced:", round(click_ratio, 4))

# 3. Interaction sparsity must be high (> 0.9 expected)
assert interaction_sparsity > 0.9, \
    "Interaction matrix not sufficiently sparse; unexpected behavior."

print("TEST-3 pass : Interaction matrix is highly sparse:",
      round(interaction_sparsity, 6))

# 4. Co-occurrence sparsity should also be very high
# Recommender datasets are typically extremely sparse (>90% empty).
# This threshold validates expected structural sparsity.
assert co_occurrence_sparsity > 0.9, \
    "Item co-occurrence sparsity unexpectedly low."

print("TEST-4 pass : Item co-occurrence is highly sparse:",
      round(co_occurrence_sparsity, 6))


# 5. Cold-start users should exist
assert cold_users > 0, \
    "No cold-start users detected; unexpected dataset behavior."

print("TEST-5 pass : Cold-start users detected:",
      cold_users)


# 6. Timestamps must be non-decreasing within each user
time_valid = behaviors.groupby("user_id")["time"].apply(
    lambda x: (x.diff().dropna() >= pd.Timedelta(0)).all()
)

assert time_valid.all(), \
    "Timestamp ordering violated within user."

print("TEST-6 pass : User timestamps are correctly ordered.")


# 7. All clicked news IDs must exist in news table
clicked_set = set(clicked_news)
news_set = set(news["news_id"])

assert clicked_set.issubset(news_set), \
    "Some clicked news IDs missing from news metadata."

print("TEST-7 pass : All clicked news IDs exist in news metadata.")

print("\nAll validation checks passed successfully.")
```

Exploratory Data Analysis

Following procedure has been followed to perform EDA after dataset loading

### 1. Data Basics  
**Reason:** To understand dataset scale and computational feasibility.

Dataset dimensions, number of unique users, and number of unique articles were computed to characterize structural size and interaction density.

---

### 2. Data Cleaning and Integrity Checks  
**Reason:** Real-world logs may contain missing values, duplicates, or malformed entries that can affect downstream analysis.

Missing values and duplicate identifiers were inspected to ensure structural consistency.

---

### 3. Impression Structure and Basket Size Analysis  
**Reason:** Impression size determines exposure complexity and affects ranking difficulty and candidate competition.

The number of candidate news items per impression was computed. This is same as basket size in transactional data and informs ranking model design. User activity distribution was also analyzed and cold-start users too.

---

### 4. Click Imbalance and Frequency of Top Items  
**Reason:** Implicit feedback datasets are typically highly imbalanced, which affects evaluation metrics and model training dynamics.

Click labels were parsed to measure overall click ratio and article popularity distribution. This step identifies long-tail behavior and engagement concentration.

---

### 5. Interaction and Co-Occurrence Sparsity  
**Reason:** Sparsity determines the suitability of matrix factorization, embedding models, and frequent pattern mining techniques.

User-item interaction sparsity was computed to quantify matrix density. Item co-occurrence sparsity was estimated to assess feasibility of association rule or frequent itemset mining approaches.

---

### 6. Bias Diagnostics  
**Reason:** Structural imbalances in exposure and content distribution can introduce systematic bias in recommendation systems.

Category frequency was analyzed to detect content imbalance. Click position distribution was examined to quantify exposure bias caused by ranking order.

---

### 7. Temporal Interaction Patterns  
**Reason:** Temporal gaps between interactions reveal burstiness and session behavior, motivating sequential or time-aware models.

Timestamps were converted and sorted to compute time gaps between user impressions, enabling analysis of engagement frequency and recency effects.

---

### 8. Correlation of User-Level Behavioral Features  
**Reason:** Understanding relationships between engagement depth, activity frequency, and exposure complexity helps guide model selection.

Aggregated user-level features were correlated to detect structural dependencies that may influence personalization strategies.

(E) Initial Insights and Direction
==================================

Following insights are based out of key observations from EDA :

**1. Extreme Interaction Sparsity**  
The user-news interaction sparsity is 0.999908, indicating that nearly all possible user-item interactions are unobserved. Similarly, item co-occurrence sparsity (0.993806) confirms that articles rarely appear together across impressions.  

This suggests that traditional frequency-based or co-occurrence-driven methods may struggle to extract stable patterns.


**2. Severe Click Imbalance**  
The overall click ratio is 4.04%, confirming strong implicit feedback imbalance. The majority of candidate items are not clicked, which reflects real-world recommendation environments.


**3. Long-Tail Popularity Distribution**  
A small subset of articles receives disproportionately high engagement (e.g., top article with 4,316 clicks), while most items receive very few clicks.  

This indicates strong popularity bias and suggests that basic recommendation strategies may exaggerate popular content.


**4. High Cold-Start Prevalence**  
Approximately 32.77% of users have only one recorded impression.  

This highlights the importance of methods that generalize beyond user-specific historical interactions.


**5. Temporal Burstiness**  
Time-gap analysis shows irregular and clustered interaction behavior.  
User activity is negatively correlated with time gaps (-0.34), indicating that highly active users interact more frequently in short durations.
